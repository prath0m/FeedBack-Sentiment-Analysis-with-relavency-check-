```python
import matplotlib.pyplot as plt
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import numpy as np

# Initialize the Sentiment Intensity Analyzer
analyzer = SentimentIntensityAnalyzer()

def is_relevant(feedback):
    relevant_keywords = ['event', 'user', 'interface', 'customer', 'support', 'crashing', 'product',
        'feedback', 'portal', 'web', 'application', 'mobile', 'quick', 'rapid', 'submission',
        'collection', 'administration', 'privilege', 'review', 'verified', 'organization',
        'group', 'category', 'date', 'range', 'activity', 'fetch', 'analyze', 'text',
        'objectionable', 'irrelevant', 'content', 'extract', 'report', 'file', 'email',
        'insight', 'data', 'persistent', 'photos', 'health', 'education', 'livelihood',
        'environment', 'mechanism', 'pc', 'ux', 'ui', 'interface', 'efficiency', 'storage',
        'relevance', 'event-conducted', 'place', 'accessibility', 'feedback-experience',
        'automation', 'admin-users', 'user-friendly', 'intuitive', 'security', 'authentication',
        'privacy', 'real-time', 'dashboard', 'metrics', 'visualization', 'export', 'import',
        'database', 'cloud', 'integration', 'cross-platform', 'responsive', 'engagement',
        'satisfaction', 'survey', 'questionnaire', 'responses', 'moderation', 'filtering',
        'scalability', 'performance', 'backup', 'restore', 'analytics', 'trends', 'summary',
        'overview', 'notifications', 'alerts', 'customization', 'templates', 'deployment',
        'monitoring', 'maintenance'
    ]

    irrelevant_keywords = ['spam', 'advertisement', 'promotion', 'unsubscribe', 'buy', 'deal', 'discount', 'name', 'fuck']

    feedback_lower = feedback.lower()
    
    if any(keyword in feedback_lower for keyword in irrelevant_keywords):
        return False

    return any(keyword in feedback_lower for keyword in relevant_keywords)

def classify_feedback(feedback):
    if is_relevant(feedback):
        sentiment_score = analyzer.polarity_scores(feedback)
        sentiment = 'Positive' if sentiment_score['compound'] > 0.05 else 'Negative' if sentiment_score['compound'] < -0.05 else 'Neutral'
        return f"Relevant feedback. Sentiment: {sentiment}"
    else:
        return "Irrelevant feedback."

feedbacks = [
    "The app is user-friendly.",
    "I don't like the taste of this restaurant's food.",
    "I had issues logging in yesterday.",
    "The weather is so nice today.",
    "the Event was too bad",
    "it was a great experience to use that app",
    "Event was excellent.",
    "App the was not giving output",
    "Check out our new discounts and promotions!"
]

# Initialize sentiment counts for relevant feedbacks only
sentiment_counts = {
    'Positive': [],
    'Neutral': [],
    'Negative': [],
}

positive_count = 0
neutral_count = 0
negative_count = 0

# Classify feedback and count sentiments
for feedback in feedbacks:
    result = classify_feedback(feedback)
    print(f"Feedback: '{feedback}' -> {result}")
    
    if "Relevant feedback" in result:
        if "Positive" in result:
            positive_count += 1
        elif "Negative" in result:
            negative_count += 1
        elif "Neutral" in result:
            neutral_count += 1
        
        # Append counts to the lists after each relevant feedback
        sentiment_counts['Positive'].append(positive_count)
        sentiment_counts['Neutral'].append(neutral_count)
        sentiment_counts['Negative'].append(negative_count)

# Convert sentiment counts to numpy arrays for calculations
positive_counts = np.array(sentiment_counts['Positive'])
neutral_counts = np.array(sentiment_counts['Neutral'])
negative_counts = np.array(sentiment_counts['Negative'])

# Calculate statistics for the control chart
mean_positive = np.mean(positive_counts)
mean_neutral = np.mean(neutral_counts)
mean_negative = np.mean(negative_counts)

std_positive = np.std(positive_counts)
std_neutral = np.std(neutral_counts)
std_negative = np.std(negative_counts)

# Define control limits
ucl_positive = mean_positive + 3 * std_positive
lcl_positive = max(0, mean_positive - 3 * std_positive)  # LCL cannot be negative
ucl_neutral = mean_neutral + 3 * std_neutral
lcl_neutral = max(0, mean_neutral - 3 * std_neutral)
ucl_negative = mean_negative + 3 * std_negative
lcl_negative = max(0, mean_negative - 3 * std_negative)

# Create the control chart
plt.figure(figsize=(12, 8))

# Plot positive sentiment counts
plt.plot(sentiment_counts['Positive'], marker='o', linestyle='-', color='g', label='Positive Count')
plt.axhline(mean_positive, color='g', linestyle='--', label='Positive Mean')
plt.axhline(ucl_positive, color='g', linestyle=':', label='Positive UCL')
plt.axhline(lcl_positive, color='g', linestyle=':', label='Positive LCL')

# Plot neutral sentiment counts
plt.plot(sentiment_counts['Neutral'], marker='o', linestyle='-', color='b', label='Neutral Count')
plt.axhline(mean_neutral, color='b', linestyle='--', label='Neutral Mean')
plt.axhline(ucl_neutral, color='b', linestyle=':', label='Neutral UCL')
plt.axhline(lcl_neutral, color='b', linestyle=':', label='Neutral LCL')

# Plot negative sentiment counts
plt.plot(sentiment_counts['Negative'], marker='o', linestyle='-', color='r', label='Negative Count')
plt.axhline(mean_negative, color='r', linestyle='--', label='Negative Mean')
plt.axhline(ucl_negative, color='r', linestyle=':', label='Negative UCL')
plt.axhline(lcl_negative, color='r', linestyle=':', label='Negative LCL')

# Add titles and labels
plt.title('Control Chart for Feedback Sentiment Analysis')
plt.xlabel('Feedback Number')
plt.ylabel('Cumulative Sentiment Count')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
